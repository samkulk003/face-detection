```python
import cv2
import matplotlib.pyplot as plt
```

```python
image=cv2.imread('C:\\Users\\Shekhar Kulkarni\\Documents\\pic.jpg')
print(image)
plt.imshow(image)
plt.show()
```

```python
face_detector=cv2.CascadeClassifier('C:\\Users\\Shekhar Kulkarni\\Documents\\haarcascade_frontalface_default.xml')
face_detector
```

```python
detections=face_detector.detectMultiScale(image,scaleFactor=1.29,minNeighbors=2,minSize=(20,20), maxSize=(100,100))
print(detections)
len(detections)
```

```python
image=cv2.imread('C:\\Users\\Shekhar Kulkarni\\Documents\\pic.jpg')
plt.imshow(image)
plt.show()
for (x,y,w,h) in detections:
    cv2.rectangle(image,(x,y),(x+w,y+h),(0,255,0),1)
plt.imshow(image)
plt.show()
```
